```python
from pyspark.sql.functions import sum

daily_sales_gold_df = (
    spark.table("retailx_catalog.silver.orders")
    .groupBy("ORDER_DATE")
    .agg(sum("AMOUNT").alias("TOTAL_SALES"))
)

daily_sales_gold_df.write.mode("overwrite") \
.saveAsTable("retailx_catalog.gold.daily_sales")
```

```python
%sql
SELECT * FROM retailx_catalog.gold.daily_sales;
```

```python
%sql
SELECT SUM(AMOUNT) AS TOTAL_AMOUNT
FROM retailx_catalog.silver.orders;
```
