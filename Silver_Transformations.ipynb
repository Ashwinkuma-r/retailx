```python
from pyspark.sql.functions import col, expr
orders_silver_df = (
    spark.table("retailx_catalog.bronze.orders")
    .withColumn(
        "ORDER_DATE",
        expr("try_to_date(ORDER_DATE, 'dd-MM-yyyy')")
    )
    .withColumn("AMOUNT", col("AMOUNT").cast("double"))
    .filter(col("ORDER_DATE").isNotNull())
    .filter(col("AMOUNT").isNotNull())
    .filter(col("AMOUNT") > 0)
)

orders_silver_df.write.mode("overwrite") \
.saveAsTable("retailx_catalog.silver.orders")
```

```python
%sql
SELECT * FROM retailx_catalog.silver.orders;
```

```python
from pyspark.sql.functions import col, expr
customers_silver_df = (
    spark.table("retailx_catalog.bronze.customers")
    .withColumn(
        "CREATED_DATE",
        expr("try_to_date(CREATED_DATE, 'dd-MM-yyyy')")
    )
    .filter(col("EMAIL").isNotNull())
    .filter(col("EMAIL").contains("@"))
    .filter(col("CREATED_DATE").isNotNull())
)

customers_silver_df.write.mode("overwrite") \
.saveAsTable("retailx_catalog.silver.customers")
```

```python
%sql
SELECT * FROM retailx_catalog.silver.customers;
```
